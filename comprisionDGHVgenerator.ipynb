## convert a positive integer x into binary

```python
def digits(x): #always binary
            x = int(x)
            le = list('{0:0b}'.format(x))
            le.reverse()
            return le
```

## Converts a positive integer x into binary with l digits

```python
def toBinary(x,l):
        x=int(x)
        return digits(x+2**l)[:-1]
```

# convert list to string

```python
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele   
    
    # return string   
    return str1 
```

# comput num to mod $2^k$ binarry 

```python
def extractKBits(num,k): 
  
     binary = toBinary(num,k) 
     binary.reverse()
     binary=binary[-k:]
     binary=listToString(binary)
     kBitSubStr=int(binary,2)
     return kBitSubStr
```

# comput num to mod $2^k$ binarry  LIST

```python
def extractKBitsLIST(num,k): 
     L=[]
     # convert number into binary first 
     for i in num:
               binary = toBinary(i,k) 
               binary.reverse()
               binary=binary[-k:]
               binary=listToString(binary)
               kBitSubStr=int(binary,2)
               L.append(kBitSubStr)
     return L 
```

## $\lfloor \frac{a}{b} \rceil$   Gives the nearest integer to $\frac{a}{b}$

```python
def QuotientNear(a,b):
    a=int(a)
    b=int(b)
    return ((a<<1)+b)//(b<<1)
```

## $ a \quad mod \quad b = a - b \lfloor \frac{a}{b} \rceil $  Computes a mod b with $a \in [\frac{-b}{2},\frac{b}{2}]$

```python
def modNear(a,b):
    return a-b*QuotientNear(a,b)
```

## $ a \quad mod \quad b = a - b \lfloor \frac{a}{b} \rceil $  Computes a mod b with $a \in [\frac{-b}{2},\frac{b}{2}]$ a=list ,b=number

```python
def mNear(a,b):
    e=[]
    for i in a:
        e.append(QuotientNear(i,b))
       # print(e)
    return np.subtract(a,np.multiply(b,e))
```

#  pseudo-random number generator

```python
def grandom(se,T,γ):
    random.seed(se)
    for i in range(T):
        yield  random.randint(0,2**γ)
    return 
```

# encryption

```python
def enc(m,se,x_0,γ,τ,delta,ρρ,α,gri,epep,gep):
    m = list(m)
    XX=grandom(se,τ,γ)
    def X():
        for i in XX:
            yield np.subtract(i,delta.__next__())
        return
    gX=X()
    XXXX=list(gX)
    #print(XXXX)
    #X=np.subtract(XX,delta)
    def b():
        for i in range(len(m)):
            b=[]
            for i in range (0,τ):
                b.append(random.randint(0,2**α))
            yield b
        return
    gb=b()
            
    c=[]
    for i in m:
                r=random.randint(-2**ρρ,2**ρρ)
                cc=np.inner(gb.__next__(),XXXX)
                ccc=cc<<1
                rr=r<<1
                ccc=i+ccc+rr
                cccc=modNear(ccc,x_0)
                c.append(cccc)
    #print("ciphertext",c)
    return c
```

# ADD multi bit

```python
def ADD(c,p,m,x_0):
    c=list(c)
    m=list(m)
    sum_addc=sum(c)
    sum_addc=modNear(sum_addc,x_0)
    sum_addc=([sum_addc])
    dc=dec(sum_addc,p)
    print("saddc",dc)
    sum_addm=sum(m)
    sum_addm=sum_addm%2
    sum_addm=([sum_addm])
    print("saddm",sum_addm)
    print(error(dc,sum_addm))
```

# MULTIPLACTION multi bit

```python
def mult(c,p,m,x_0):
    c=list(c)
    m=list(m)
    mul_c=np.prod(c)
    print(mul_c)
    mul_c=modNear(mul_c,x_0)
    mul_c=list([mul_c])
    dc=dec(mul_c,p)
    print("mullc",dc)
    mul_m=np.prod(m)
    mul_m=mul_m%2
    mul_m=list([mul_m])
    print("mul_m",mul_m)
    print(error(dc,mul_m))   
```
